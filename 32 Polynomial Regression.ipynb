##### Polynomial Regression

```python
df = pd.read_csv('datasets/Salary_Data.csv')
x = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
print('Accuracy:', reg.score(x,y))
```

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
```

```python
x_poly
```

```python
reg.fit(x_poly, y)
print('Accuracy:', reg.score(x_poly, y))
```

```python
poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(x)

reg.fit(x_poly, y)
print('Accuracy:', reg.score(x_poly, y))
```

```python
poly = PolynomialFeatures(degree=5)
x_poly = poly.fit_transform(x)

reg.fit(x_poly, y)
print('Accuracy:', reg.score(x_poly, y))
```

```python
poly = PolynomialFeatures(degree=6)
x_poly = poly.fit_transform(x)

reg.fit(x_poly, y)
print('Accuracy:', reg.score(x_poly, y))
```

```python
acc = []
for i in range(2,21):
    poly = PolynomialFeatures(degree=i)
    x_poly = poly.fit_transform(x)

    reg.fit(x_poly, y)
    acc.append(reg.score(x_poly, y)) 
```

```python
acc
```

```python
acc = []
for i in range(2,31):
    poly = PolynomialFeatures(degree=i)
    x_poly = poly.fit_transform(x)

    reg.fit(x_poly, y)
    acc.append(reg.score(x_poly, y)) 
```

```python
acc
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xticks(range(2,31))
plt.xlabel('Degree')
plt.ylabel('Accuracy')
plt.plot(range(2,31), acc, color ='red')
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xticks(range(2,31))
plt.xlabel('Degree')
plt.ylabel('Accuracy')
plt.plot(range(2,31), acc, color ='red', lw = 2)
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xticks(range(2,31))
plt.xlabel('Degree')
plt.ylabel('Accuracy')
plt.plot(range(2,31), acc, color ='r', lw = 2)
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xticks(range(2,31))
plt.xlabel('Degree')
plt.ylabel('Accuracy')
plt.plot(range(2,31), acc, color ='r', lw=2, marker='o')
```

```python
poly = PolynomialFeatures(degree=12)
x_poly = poly.fit_transform(x)

reg.fit(x_poly, y)
reg.score(x_poly, y)
```

```python
reg.coef_
```

```python
y_pred = reg.predict(x_poly)
```

```python
plt.grid()
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(x, y, color = 'green', label = 'actual')
plt.plot(x, y_pred, color = 'red', label = 'predicted')
plt.legend()
```
