##### Standardization

```python
x = np.array([500,800,300,600,900,700,400])
```

```python
def standardize(x):
    x_new = (x - x.mean()) / x.std()
    return x_new
```

```python
x_new = standardize(x)
```

```python
x_new
```

```python
x_new.std()
```

```python
x_new.mean()
```
