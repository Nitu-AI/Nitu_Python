```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
# Sobel parameters: ksize, dx, dy, ddepth
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
# Sobel for horizontal edges
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
# Sobel for vertical edges  

# Combine the Sobel images (optional)
combined_sobel = cv2.add(sobelx, sobely)
```

```python
final_img = np.hstack([sobelx, sobely, combined_sobel])
cv2.imshow('All Images', final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,4,1);
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.subplot(1,4,2); 
plt.title('Sobel X')
plt.imshow(sobelx, cmap='gray')
plt.subplot(1,4,3)
plt.title('Sobel y')
plt.imshow(sobely, cmap='gray')
plt.subplot(1,4,4)
plt.title('Final')
plt.imshow(combined_sobel, cmap='gray')
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,4,1);
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.subplot(1,4,2); 
plt.title('Sobel X')
plt.imshow(abs(sobelx), cmap='gray')
plt.subplot(1,4,3)
plt.title('Sobel y')
plt.imshow(abs(sobely), cmap='gray')
plt.subplot(1,4,4)
plt.title('Final')
plt.imshow(abs(combined_sobel), cmap='gray')
```

##### Canny Edge Detection

```python
import cv2

# Load an image
img = cv2.imread('images/lenna.png') # Load as grayscale

# Apply Canny edge detection
edges = cv2.Canny(img, 50, 150) # Adjust thresholds if needed

# Display results
cv2.imshow('Original Image', img)
cv2.imshow('Canny Edge Detection', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()                
```

```python
import cv2

# Load an image
img = cv2.imread('images/lenna.png') # Load as grayscale

# Apply Canny edge detection
edges = cv2.Canny(img, 50, 150) # Adjust thresholds if needed

# Display results
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2)
plt.title('Edges Defected')
plt.imshow(edges, cmap='gray')  
```
