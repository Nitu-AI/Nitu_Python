```python
# Create tensors
x = tf.constant([[1,2,3],[7,8,9]])
y = tf.constant([[4,5],[6,2]])

print("Tensor a:",x)
print("Tensor b:",y)
```

```python
# Create tensors
x = tf.constant([[1,2,3],[7,8,9]], dtype=np.int8)
y = tf.constant([[4,5],[6,2]])

print("Tensor a:",x)
print("Tensor b:",y)
```

```python
# Create tensors
x = tf.constant([[[1,2,3],[7,8,9]],[[0,2,5],[7,6,8]]], dtype=np.int8)
y = tf.constant([[4,5],[6,2]])

print("Tensor a:",x)
print("Tensor b:",y)
```
